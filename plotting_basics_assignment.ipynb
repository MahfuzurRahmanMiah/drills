```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


df = pd.read_csv('oil_price_data.csv')

def remove_sign(list, str):
    newlist = []
    for x in list:
        y = x.replace(str, "")
        newlist.append(float(y))
    return newlist

#print(df)

df['Nominal Price as Int'] = remove_sign(df['Nominal Price'], "$")
df['Inflation Adjusted Gas Price as Int'] = remove_sign(df['Inflation Adjusted Gas Price'],"$")
print(df)

```

Let's make a dataframe variable, df, from the simple file called oil_price_data.csv using the read_csv() function. We also need to remove the $ sign from the dollar values and we'd like the values to be floats for easy processing. 

```python
x = df['Date']
y = df['Nominal Price as Int']
plt.plot(x, y, color='green', linestyle='dashed', marker='h',
     markerfacecolor='blue', markersize=2)
plt.ylabel('Cost in Dollars ($)')
plt.xlabel('Year (CE)')
plt.title('Historical Data on Gas Prices')

plt.show()
```

Here we see the price per gallon of gas in the US over the last century. We note that there is a clear inflection point in the 1970s with oil prices never returning to prices below \$0.50. A small bump is noticed in 1973 followed by a spike in prices in 1979, inline with the oil crises of 1973 and 1979. Prices in the last decade have volleyed between \$2.00 and \$4.00.


```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
x1 = df['Date']
y1 = df['Nominal Price as Int']
plt.plot(x1, y1, color='green', linestyle='dashed', marker='h',
     markerfacecolor='blue', markersize=2)
plt.ylabel('Cost in Dollars ($)')
plt.xlabel('Year (CE)')
plt.title('Historical Data on Gas Prices')

plt.subplot(1, 2, 2)
x2 = df['Date']
y2 = df['Inflation Adjusted Gas Price as Int']
plt.plot(x2, y2, color='red', linestyle='solid', marker='o',
     markerfacecolor='orange', markersize=2)
plt.ylabel('Cost in Dollars ($)')
plt.xlabel('Year (CE)')
plt.title('Historical Data on Gas Prices Adjusted for Inflation')

plt.tight_layout()
plt.show()
```

However, looks can be deceiving. While our first graph makes it seem that oil has only gotten more expensive over time, when we take into consideration inflation, we see that oil has actually become cheaper from the 1920s to the turn of the century. The last two decades has seen a return of oil valued at a high price. What happens when we overlay these graphs?

```python
plt.figure(figsize=(15, 7))

x1 = df['Date']
y1 = df['Nominal Price as Int']
plt.plot(x1, y1, color='green', linestyle='dashed', marker='h',
     markerfacecolor='blue', markersize=2)
x2 = df['Date']
y2 = df['Inflation Adjusted Gas Price as Int']
plt.plot(x2, y2, color='red', linestyle='solid', marker='o',
     markerfacecolor='orange', markersize=2)
plt.ylabel('Cost in Dollars ($)')
plt.xlabel('Year (CE)')
plt.title('Historical Data on Gas Prices Adjusted for Inflation')

plt.tight_layout()
plt.show()
```

With this view, we are able to see certain spikes in price are still reflected in the inflation adjusted prices, such as the already mentioned spikes in 1973 and 1979. What was missed was the fact that though prices seem somewhat stable between the 1980s to the turn of the century, when we take into account inflation, the value of oil was effectively decreasing during that time frame.
